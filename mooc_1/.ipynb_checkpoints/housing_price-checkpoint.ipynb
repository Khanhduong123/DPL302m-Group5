```python
data= pd.read_csv("D:\Khanh_Duong\BAO KHANH\DPL302m-Group5\housing_price\Housing.csv")
data
```

```python
nan_check = data.isnull().values.any()
print("NaN check:", nan_check)
```

```python
data = data.select_dtypes(['float64','int64','int32'])
```

```python
data= data.apply(np.log1p)
data
```

```python
sns.distplot(data['price'])
```

```python
X = data.drop(['price'],axis=1)
Y= data['price']
```

```python
X_train, X_temp, Y_train, Y_temp = train_test_split(X.iloc[:500], Y.iloc[:500], test_size=0.2, random_state=42)
X_dev, X_test, Y_dev, Y_test = train_test_split(X_temp, Y_temp, test_size=0.5, random_state=42)
```

```python
num_features = X_train.shape[1]
```

```python
model= tf.keras.models.Sequential()
model.add(tf.keras.layers.Input(shape=(num_features,)))
model.add(tf.keras.layers.Dense(5,activation='linear'))
model.add(tf.keras.layers.Dense(3,activation='linear'))
model.add(tf.keras.layers.Dense(1,activation='linear'))
model.summary()
```

```python
model.compile(optimizer='adam', loss='mean_squared_error')
```

```python
model.fit(X_train, Y_train, epochs=1000, batch_size=32)
```

```python
test_loss = model.evaluate(X_test, Y_test)
```

```python
predictions = model.predict(X_test)
predictions
```
