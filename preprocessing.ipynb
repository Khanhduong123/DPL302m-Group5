```python
import numpy as np
import pandas as pd
```

```python
phone_data= pd.read_csv('./Datas/thongtindienthoai.csv')
laptop_data = pd.read_csv('./Datas/thongtinlaptop.csv')
```

```python
updated_phone_names = []
for i in range(len(phone_data['phone_names'])):
    new_data = phone_data['phone_names'][i].replace(phone_data['phone_ids'][i], "")
    updated_phone_names.append(new_data)
phone_data['phone_names'] = updated_phone_names
```

```python
new_names=[]
for i in laptop_data['laptop_names']:
    new_name =i.split()[0:4]
    new_name = ' '.join(new_name)
    new_names.append(new_name)
laptop_data['laptop_names']= new_names
```

```python
# phone_data.drop(['phone_ids'],axis=1,inplace=True)
laptop_data.drop(['laptop_ids'],axis=1,inplace=True)
```
