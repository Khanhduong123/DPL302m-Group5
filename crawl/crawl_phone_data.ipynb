```python
driver_1= webdriver.Chrome(service=service,options=chrome_options)
driver_2= webdriver.Chrome(service=service,options=chrome_options)
url_1= 'https://fptshop.com.vn/dien-thoai?sort=ban-chay-nhat&trang=4'
url_2= 'https://fptshop.com.vn/may-tinh-xach-tay'
driver_1.get(url_1)
driver_2.get(url_2)
```

```python
#Scrape the URLs of the profiles
#extract the URLs of one page
def GetURL_for_phone():
    page_source = BeautifulSoup(driver_1.page_source)
    profiles= page_source.find_all('a',class_="cdt-product__name")
    all_URLs_profile = []
    for profile in profiles:
        profile_ID= profile.get('href')
        profile_URL= 'https://fptshop.com.vn/' + profile_ID
        if profile_URL not in all_URLs_profile:
            all_URLs_profile.append(profile_URL)
    return all_URLs_profile

def GetURL_for_laptop():
    page_source = BeautifulSoup(driver_2.page_source)
    profiles= page_source.find_all('a',class_="cdt-product__name")
    all_URLs_profile = []
    for profile in profiles:
        profile_ID= profile.get('href')
        profile_URL= 'https://fptshop.com.vn/' + profile_ID
        if profile_URL not in all_URLs_profile:
            all_URLs_profile.append(profile_URL)
    return all_URLs_profile
```
